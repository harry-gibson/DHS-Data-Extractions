```python
all_data = pd.read_csv('raw_child_mortality.csv', low_memory=False, 
                       usecols=['surveyid','clusterid', 'sample_wt', 'v008_int_cmc', 'b3_dob_cmc', 'b5_is_alive', 'b7_death_age_months']
                      )
```

```python
all_data.drop(all_data[all_data['surveyid']!=491].index, inplace=True)
```

```python
pd.set_option('display.max_columns', None)
```

```python
all_data[~all_data.b7_death_age_months.isna()]
```

```python
age_classes = {
    #1: (0,1)     # NN
     #,2: (1,12)   # infant after NN
    #,3: (12,60)  #child after infancy
    #,4: (0, 12)  # infant
    5: (0, 60)  # U5
}
age_int_ids = [5
               #,2
               #,3
               #,4
               #, 5
              ] # just to easily iterate in a fixed order

time_windows = (5,10)
#i_1  = (0,1,1)
#i_2 = (1,12,11)
#i_3 = (12,60,48)
```

```python
for w in time_windows:
    all_data[f'start_mth_{w}yr'] = all_data['v008_int_cmc'] + 12 * (-w + 1) - 12
    all_data[f'end_mth_{w}yr'] = all_data['v008_int_cmc'] + 12 * 0 - 1
    #all_data[f'maybe_in_{w}yr'] = (
    #    (all_data.b3_dob_cmc <= all_data[f'end_mth_{w}yr']) & 
    #    (all_data.b3_dob_cmc + 60 >= all_data[f'start_mth_{w}yr']-1))
```

```python
all_data[~all_data.b7_death_age_months.isna()]
```

```python
all_data['age_int_at_death']=np.nan
for ageclass in age_int_ids:
    min_age, max_age = age_classes[ageclass]
    all_data.loc[(all_data.b7_death_age_months>=min_age) & (all_data.b7_death_age_months<max_age), 
                 'age_int_at_death'] = ageclass
```

```python
all_data[~all_data.b7_death_age_months.isna()]
```

```python
for w in time_windows:
    for ageclass in age_int_ids:
        min_age, max_age = age_classes[ageclass]
        all_data.loc[all_data.age_int_at_death==ageclass, 
                     f'dod_{w}yr_from'] = all_data.b3_dob_cmc + min_age
        all_data.loc[all_data.age_int_at_death==ageclass, 
                     f'dod_{w}yr_to'] = all_data.b3_dob_cmc + max_age
    all_data.loc[all_data[f'dod_{w}yr_to'] < all_data[f'start_mth_{w}yr'], 
                 [f'dod_{w}yr_from',f'dod_{w}yr_to']]=np.nan
```

```python
all_data[~all_data.b7_death_age_months.isna()]
```

```python
# Count dead children whose age was in the given interval ONLY at the time of death and where this 
# occurred within the time window of interest
# DHS lines 242-249
for w in time_windows:
    for ageclass in age_int_ids:
        min_age, max_age = age_classes[ageclass]

        # window: five years before survey
        #  died in this age interval, and for this child that age interval is entirely in the time window, count death as 1
        all_data.loc[(all_data.age_int_at_death==ageclass) & 
                     (all_data[f'dod_{w}yr_to']<=all_data[f'end_mth_{w}yr']) & 
                     (all_data[f'dod_{w}yr_from']>=all_data[f'start_mth_{w}yr']), 
                     f'died_{w}yr_{ageclass}'] = 1
        # age interval is partly in the time window and partly before it, count death as 0.5
        all_data.loc[(all_data.age_int_at_death==ageclass) & 
                     (all_data[f'dod_{w}yr_to']>=all_data[f'start_mth_{w}yr']) & 
                     (all_data[f'dod_{w}yr_from']<all_data[f'start_mth_{w}yr']), 
                     f'died_{w}yr_{ageclass}'] = 0.5
        # age interval is partly in the time window and partly after it, count death as 0.5
        all_data.loc[(all_data.age_int_at_death==ageclass) & 
                     (all_data[f'dod_{w}yr_to']>all_data[f'end_mth_{w}yr']) & 
                     (all_data[f'dod_{w}yr_from']<=all_data[f'end_mth_{w}yr']), 
                     f'died_{w}yr_{ageclass}'] = 0.5

        # window: ten years before survey
        #  age interval is entirely in the time window, count death as 1
#        all_data.loc[(all_data.age_int_at_death==ageclass) & 
#                     (all_data.dod_10yr_to<=all_data.end_mth_10yr) & 
#                     (all_data.dod_10yr_from>=all_data.start_mth_10yr), 
#                     f'died_10yr_{ageclass}'] = 1
        # age interval is partly in the time window and partly before it, count death as 0.5
#        all_data.loc[(all_data.age_int_at_death==ageclass) & 
#                     (all_data.dod_10yr_to>=all_data.start_mth_10yr) & 
#                     (all_data.dod_10yr_from<all_data.start_mth_10yr), 
#                     f'died_10yr_{ageclass}'] = 0.5
        # age interval is partly in the time window and partly after it, count death as 0.5
#        all_data.loc[(all_data.age_int_at_death==ageclass) & 
#                     (all_data.dod_10yr_to>all_data.end_mth_10yr) & 
#                     (all_data.dod_10yr_from<=all_data.end_mth_10yr), 
#                     f'died_10yr_{ageclass}'] = 0.5
```

```python
all_data[~all_data.b7_death_age_months.isna()]
```

```python
# NOW DO RISK i.e. denominators
# All children alive or dead whose age was in the given interval OR OLDER within the time window
for w in time_windows:
    for ageclass in age_int_ids:
        min_age, max_age = age_classes[ageclass]
        # DHS lines 254-267

        # window: five years before survey
        #  age interval is entirely in the time window, count risk as 1 #L254
        all_data.loc[((all_data.age_int_at_death>=ageclass) | all_data.b7_death_age_months.isna()) &
                    (all_data.b3_dob_cmc + max_age <= all_data[f'end_mth_{w}yr']) &
                    (all_data.b3_dob_cmc + min_age >= all_data[f'start_mth_{w}yr']),
                    f'risk_{w}yr_{ageclass}'] = 1
        #L257 
        # NB DHS has "age_int_at_death>`i'" which i think is wrong, it should be "age_int_at_death>=`i'"
        all_data.loc[((all_data.age_int_at_death>=ageclass) | all_data.b7_death_age_months.isna()) &
                    (all_data.b3_dob_cmc + max_age >= all_data[f'start_mth_{w}yr']) &
                    (all_data.b3_dob_cmc + min_age < all_data[f'start_mth_{w}yr']),
                    f'risk_{w}yr_{ageclass}'] = 0.5
        #L260
        all_data.loc[((all_data.age_int_at_death>=ageclass) | all_data.b7_death_age_months.isna()) &
                    (all_data.b3_dob_cmc + max_age > all_data[f'end_mth_{w}yr']) &
                    (all_data.b3_dob_cmc + min_age <= all_data[f'end_mth_{w}yr']),
                    f'risk_{w}yr_{ageclass}'] = 0.5

        # L271
        all_data.loc[all_data[f'died_{w}yr_{ageclass}'] == 0.5, 
                    f'risk_{w}yr_{ageclass}'] = 0.5

        # L274
        all_data.loc[(all_data[f'died_{w}yr_{ageclass}'].isna()) & 
                     (all_data[f'risk_{w}yr_{ageclass}'] > 0) &
                     (all_data[f'risk_{w}yr_{ageclass}'] <= 1), 
                    f'died_{w}yr_{ageclass}'] = 0
        
        # L275
        all_data.loc[(all_data[f'risk_{w}yr_{ageclass}'].isna()) & 
                     (all_data[f'died_{w}yr_{ageclass}'] > 0) &
                     (all_data[f'died_{w}yr_{ageclass}'] <= 1), 
                    f'risk_{w}yr_{ageclass}'] = 0
        
        all_data.loc[all_data[f'died_{w}yr_{ageclass}'] == 1, 
                    f'risk_{w}yr_{ageclass}'] = 1
```
